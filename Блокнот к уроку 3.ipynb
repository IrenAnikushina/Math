```python
#1.	Напишите код, который будет переводить полярные координаты в декартовы.

def polar_to_dec(x,y):
    R = sqrt(x**2 + y**2)
    A = np.arctan2(y,x)
    X = R*np.cos(A)
    Y = R*np.sin(A)
    
    print(f'R = ', round(R, 2))
    print(f'A = ', round(A, 2), 'рад')
    print(f'Полярные координаты: (', round(A*180/np.pi, 2),',',round(R, 2),')')
    print(f'Проверка! Изначальные координаты точки:', '(',int(X),',',int(Y),')')
```

```python
polar_to_dec(5,6)
```
